```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

from sklearn.datasets import make_classification
```

Create Time Series data

```python
entities = [ "E" + str(i) for i in range(1,11)]
entities
```

```python
def generate_data_for_entity(entity_name : str, start_year:int, n_years:int) -> pd.core.frame.DataFrame:

  X, y = make_classification(n_samples=n_years * 365, n_features=10, n_classes=10, random_state=2022, n_informative= 5)

  df = pd.DataFrame(X, columns=["f" + str(i) for i in range(1,11)])
  df["y"]  = y
  df["entity"] = entity_name

  dt = datetime(start_year - 1,12,31)
  days_rating =[]
  for _ in range(n_years * 365):
    dt = dt + timedelta(days = 1)
    days_rating.append(dt)

  df["date"]= days_rating

  final_columns = ["date", "entity"] + ["f" + str(i) for i in range(1,11)] + ["y"]

  df = df[final_columns]

  return df
```

```python
dfs = []
for e in entities:
  dfs.append(generate_data_for_entity(e, 2015, 4))
```

```python
df = pd.concat(dfs, axis=0)
```
